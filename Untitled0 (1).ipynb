```python
import PyPDF2

# Load your PDF file and extract text
pdf_path = 'unit-1-renewable-energy-resource.pdf'
with open(pdf_path, 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    text = ""
    for page_num in range(len(reader.pages)):
        page = reader.pages[page_num]
        text += page.extract_text()

print(text[:1000])  # Print first 1000 characters for verification
```

```python
def split_text(text, chunk_size=500, overlap=50):
    words = text.split()
    chunks = []
    for i in range(0, len(words), chunk_size - overlap):
        chunk = " ".join(words[i:i + chunk_size])
        chunks.append(chunk)
    return chunks

documents = split_text(text)
print(f"Number of chunks created: {len(documents)}")
print(documents[:2])  # Display the first few chunks
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Load the sentence-transformer model
embedder = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Generate embeddings for each document chunk
embeddings = embedder.encode(documents)
print("Embedding shape:", embeddings.shape)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Load the sentence-transformer model
embedder = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Generate embeddings for each document chunk
embeddings = embedder.encode(documents)
print("Embedding shape:", embeddings.shape)
```

```python
import faiss

# Initialize FAISS index
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))

# Function to retrieve top-k similar documents
def retrieve(query, k=3):
    query_embedding = embedder.encode([query])
    distances, indices = index.search(query_embedding, k)
    return [documents[idx] for idx in indices[0]]
```
